## Creating a New column for flag which basically indicates whether the Order is cancelled or based on Meta-Data

```python
def FirstLetterStartsWithC(x):
    if str(x).startswith('C'):
        return 'Order Cancelled'
    else:
        return 'Order Not Cancelled'

#df['Order Cancelled Flag'] 
df['Order Cancelled Flag'] = df['Invoice'].apply(FirstLetterStartsWithC)
```
