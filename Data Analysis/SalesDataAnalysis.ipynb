```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv(
                r'C:\Users\badri\OneDrive\Desktop\Retail Data Analysis\Data__Files\Sales__cleaned.csv',
                 index_col=0
                )
df.head(5)
```

```python
df['Customer ID'] = df['Customer ID'].astype(str)
```

```python
df.dtypes
```

```python
df['Customer ID'] = df['Customer ID'].apply(lambda x: str(x).split('.')[0])
```

```python
df.head(5)
```

```python
df['Country'].replace({'RSA': 'South Africa',
                       'EIRE': 'Ireland'}, inplace=True)
```

```python
df.groupby('Country').agg(TotalNumberOfOrdersPlace = ('Customer ID','count')).sort_values(by=['TotalNumberOfOrdersPlace'],ascending=False).reset_index()
```

```python
df.groupby('Country').agg(TotalNumberOf_Order=('Customer ID','count'),
                       AverageRevenuePer_Order = ('Revenue','mean')
                       ).round({'AverageRevenuePer_Order':0}). \
                    sort_values(by= ['TotalNumberOf_Order'], ascending=False)
```
